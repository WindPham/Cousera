```python
import pandas as pd
import numpy as np
import csv
import os
import sqlite3 as sql
```

```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDS0321ENSkillsNetwork26802033-2021-01-01'
```

```python
df = pd.read_csv(url);
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%d-%m-%Y')
```

```python
def sql_to_df(cursor):
  res_df = None;
  if cursor.description:
    cols = [column[0] for column in cursor.description]
    res_df = pd.DataFrame.from_records(data = cursor.fetchall(), columns = cols)
  return res_df;
```

```python
conn = sql.connect("spacex.db");
```

```python
df.to_sql('SPACEXTBL', conn, if_exists='replace', index = False);
```

## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
query = "select distinct(\"Landing _outcome\") from SPACEXTBL";
pro1 = conn.execute(query);
df1 = sql_to_df(pro1);
df1
```

```python
query = "select distinct(Launch_Site) from SPACEXTBL";
pro1 = conn.execute(query);
df1 = sql_to_df(pro1);
df1
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
query = "select * from SPACEXTBL where Launch_Site like 'CCA%' limit 5";
pro2 = conn.execute(query);
df2 = sql_to_df(pro2);
df2
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
query = "select sum(PAYLOAD_MASS__KG_) as 'Total_Payload_mass' from SPACEXTBL where Customer == 'NASA (CRS)'";
pro3 = conn.execute(query);
df3 = sql_to_df(pro3);
df3
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
query = "select avg(PAYLOAD_MASS__KG_) from SPACEXTBL where Booster_version like 'F9 v1.1%'";
pro4 = conn.execute(query);
df4 = sql_to_df(pro4);
df4
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
query = "select date from SPACEXTBL where \"Landing _Outcome\" == 'Success (ground pad)' order by date asc limit 1";
pro5 = conn.execute(query);
df5 = sql_to_df(pro5);
df5
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
query = " select booster_version from SPACEXTBL where \"Landing _Outcome\"=='Success (drone ship)' \
          and PAYLOAD_MASS__KG_ between 4000 and 6000";
pro6 = conn.execute(query);
df6 = sql_to_df(pro6);
df6
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
query = "select count(mission_outcome) from SPACEXTBL group by mission_outcome";
pro7 = conn.execute(query);
df7 = sql_to_df(pro7);
df7
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
query = "select booster_version from SPACEXTBL where PAYLOAD_MASS__KG_ == (select max(PAYLOAD_MASS__KG_) from SPACEXTBL)";
pro8 = conn.execute(query);
df8 = sql_to_df(pro8);
df8
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
query = " select \"Landing _Outcome\", Booster_version, Launch_Site, date \
          from SPACEXTBL where \"Landing _Outcome\" == 'Failure (drone ship)' and \
          date between '2015-01-01' and '2015-12-31'";
pro9 = conn.execute(query);
df9 = sql_to_df(pro9);
df9
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
query = " select date, \"Landing _Outcome\" from SPACEXTBL \
          where date between '2010-06-04' and '2017-03-20' and \
          (\"Landing _Outcome\" == 'Failure (drone ship)' or \"Landing _Outcome\" == 'Success (ground pad)') \
order by date desc";
pro10 = conn.execute(query);
df10 = sql_to_df(pro10);
df10
```
